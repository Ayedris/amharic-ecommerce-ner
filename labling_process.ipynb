```python
import pandas as pd
import os
import ast

# Path to your original CSV file
input_path = r'C:\Users\ayedr\amharic-ecommerce-ner\data\raw\filtered_channels.csv'

# Load your CSV
df = pd.read_csv(input_path, encoding='utf-8')

# Function to clean tokens
def clean_tokens(token_list):
    # Remove tokens that are only dots, commas, or spaces
    return [t for t in token_list if not all(c in ['.', ',', '…', ' '] for c in t)]

# Parse string to list and clean
def parse_and_clean(token_str):
    try:
        tokens = ast.literal_eval(token_str)
        cleaned = clean_tokens(tokens)
        return cleaned
    except:
        return []

# Apply to DataFrame
df['clean_tokens'] = df['tokens'].apply(parse_and_clean)

# Save to new file
output_path = os.path.join(os.path.dirname(input_path), 'filtered_channels_cleaned.csv')
df.to_csv(output_path, index=False, encoding='utf-8')

print(f"Cleaned data saved to {output_path}")
```

```python
import ast

# Assuming your DataFrame 'df' is already loaded
labels_list = []

for idx, row in df.iterrows():
    tokens = row['clean_tokens']
    if isinstance(tokens, str):
        tokens = ast.literal_eval(tokens)
    
    labels = []
    i = 0
    while i < len(tokens):
        token = tokens[i]
        
        # Apply rules
        if i == 0:
            labels.append('B-product')
        elif i < 3:
            labels.append('I-product')
        elif token == 'ዋጋ፦':
            labels.append('B-price')
            if i + 1 < len(tokens):
                labels.append('I-price')
                i += 1
        elif token == 'አድራሻ':
            labels.append('B-LOC')
            # Check if next token is the specific phrase
            if i + 1 < len(tokens) and tokens[i + 1] == 'መገናኛ_መሰረት_ደፋር_ሞል_ሁለተኛ_ፎቅ':
                labels.append('I-LOC')
                i += 1
        else:
            labels.append('O')
        i += 1
    labels_list.append((tokens, labels))

# Save in CoNLL format
with open('label-messages.conll', 'w', encoding='utf-8') as f:
    for tokens, labels in labels_list:
        for token, label in zip(tokens, labels):
            f.write(f"{token}\t{label}\n")
        f.write("\n")  # blank line after each message
```
